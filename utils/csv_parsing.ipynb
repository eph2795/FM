```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
```

```python
fname = '../../datasets/forestfires.csv'
```

```python
df = pd.read_csv(fname)
```

```python
X_cat = df[['month', 'day', 'X', 'Y', 'RH']]
X_f = df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'wind', 'rain']]
Y = df[['area']]
```

```python
X_cat = np.array(X_cat, dtype=np.str)
X_f = np.array(X_f)
Y = np.array(Y).flatten()
```

```python
le = LabelEncoder()
l = []
for i in range(X_cat.shape[1]):
    l.append(le.fit_transform(X_cat[:, i]).reshape(-1, 1))
X_cat = np.concatenate(l, axis=1)

ohe = OneHotEncoder(sparse=False)
X_cat = ohe.fit_transform(X_cat)
X = np.concatenate([X_f, X_cat], axis=1)
```

```python
model = SGDRegressor(penalty=None, learning_rate='constant', eta0=1e-6, max_iter=100)
```

```python
X, Y = shuffle(X, Y)
```

```python
model.fit(X, Y)
```

```python
model.coef_
```

```python
pred = model.predict(X)
```

```python
for i in range(pred.shape[0]):
    print(pred[i], Y[i])
```

```python
(np.square(pred - Y)).mean()
```
