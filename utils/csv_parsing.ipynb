```python
import pandas as pd
import numpy as np
from scipy.sparse import hstack
from sklearn.utils import shuffle
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
```

```python
fname = '../../datasets/forestfires.csv'
# fname = '../../datasets/machine.data'
# fname = '../../datasets/avazu/train.csv'
# fname = '../../datasets/u.data.csv'
```

```python
target_col = 12
```

```python
df = pd.read_csv(fname, delimiter=',')
```

```python
df
```

```python
Y = df.iloc[:, target_col]
```

```python
cols = [i for i in range(df.shape[1])]
```

```python
cols = list(set(cols).difference(set([target_col])))
```

```python
df = df.iloc[:, cols]
```

```python
X_cat = df.select_dtypes(exclude='floating')
X_f = df.select_dtypes(include='floating')
```

```python
X_cat = np.array(X_cat, dtype=np.str)
X_f = np.array(X_f)
Y = np.array(Y).flatten()
```

```python
le = LabelEncoder()
l = []
for i in range(X_cat.shape[1]):
    l.append(le.fit_transform(X_cat[:, i]).reshape(-1, 1))
X_cat = np.concatenate(l, axis=1)

ohe = OneHotEncoder()
X_cat = ohe.fit_transform(X_cat)

if X_cat.shape[1] == 0:
    X = X_f
elif X_f.shape[1] == 0:
    X = X_cat
else:
    X = hstack([X_f, X_cat], format='csr')
```

```python
X.shape
```

```python
model = SGDRegressor(penalty=None, learning_rate='constant', eta0=1e-6, max_iter=10000)
```

```python
model.fit(X, Y)
```

```python
pred = model.predict(X)
```

```python
for i in range(pred.shape[0]):
    print(pred[i], Y[i])
```

```python
(np.square(pred - Y)).mean()
```
