```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('../../movielens/ml-100k/u.data', sep='\t', header=None)
```

```python
df = df.select_dtypes(exclude=['floating']) 
```

```python
df = df.iloc[:, :3]
```

```python
df
```

```python
ohe = OneHotEncoder(sparse=False)
```

```python
ohe.fit(df)
```

```python
dense = ohe.transform(df)
```

```python
import numpy as np
```

```python
new_df = pd.DataFrame(dense, dtype=np.uint32)
```

```python
new_df.to_csv('../../movielens/ml-100k/u.data.ohe', sep='\t')
```
